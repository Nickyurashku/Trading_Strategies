```python
const = const.iloc[:, :5].copy()
const
```

```python
const.rename(columns = {'Date added':'Date_added'},inplace=True) #need to get rid of the white space in one of the columns in df
```

```python
const.Date_added = pd.to_datetime(const.Date_added)
```
